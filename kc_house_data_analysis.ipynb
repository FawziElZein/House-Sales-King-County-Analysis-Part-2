## Part 2:

```python
# The below formula was inspired from https://www.geeksforgeeks.org/program-distance-two-points-earth/

from math import radians, cos, sin, asin, sqrt
def get_distance(lat1,lon1,lat2,lon2):
    lon1 = radians(lon1)
    lon2 = radians(lon2)
    lat1 = radians(lat1)
    lat2 = radians(lat2)
    a = sin((lat2 - lat1) / 2)**2 + cos(lat1) * cos(lat2) * sin((lon2 - lon1) / 2)**2
    c = 2 * asin(sqrt(a))
    return (c * 6371)
```

```python

from math import radians, cos, sin, asin, sqrt
def get_houses_distance_2(idx,idy):
    min_id = df_yr_gt_2000['id'].min();
    max_id = df_yr_gt_2000['id'].max();
    if not min_id<idx<max_id:
        print(idx,"is not included in the data set")
        return None
    if not min_id<idy<max_id:
        print(idy,"is not included in the data set")
        return None
    
    idxrow = df.loc[df['id']==idx]
    idyrow = df.loc[df['id']==idy]
    
    print("calculating distance between ",[idxrow.iloc[0]['lat'],idxrow.iloc[0]['long']]," and ",[idyrow.iloc[0]['lat'],idyrow.iloc[0]['long']])

    return get_distance(idxrow.iloc[0]['lat'],
                        idxrow.iloc[0]['long'],
                        idyrow.iloc[0]['lat'],
                        idyrow.iloc[0]['long']
                       )
```

#### Clustering longs and lats

```python
X=df_yr_gt_2000.loc[:,['lat','long']]
X
```

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot

kmeans = KMeans(n_clusters = 3, init ='k-means++')
kmeans.fit(X[X.columns[:]]) # Compute k-means clustering.
X['cluster_label'] = kmeans.fit_predict(X[X.columns[:]])
centers = kmeans.cluster_centers_ # Coordinates of cluster centers.
labels = kmeans.predict(X) 

X.plot.scatter(x = 'lat', y = 'long',c=labels, s=3,cmap='viridis')
pyplot.scatter(x=centers[:, 0], y=centers[:, 1], c='black', s=300, alpha=0.5)
print(centers)
print(centers[:, 0])
```
